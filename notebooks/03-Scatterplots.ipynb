# Scatterplots

[Matplotlib Tutorial#Scatter Plots](http://www.labri.fr/perso/nrougier/teaching/matplotlib/#scatter-plots)

```python
# Imports
import numpy as np
import pandas as pd
import altair as alt
```

```python
# Create Data
n = 1024
x = np.random.normal(0, 1, n)
y = np.random.normal(0, 1, n)
angle = np.arctan2(y,x)
df = pd.DataFrame({'x':x, 'y':y, 'angle':angle})

# Make Chart
chart = alt.Chart(df)
color_encoding = alt.Color('angle:Q', legend=None, 
                           scale=alt.Scale(domain=[np.pi, -np.pi], 
                                           scheme='spectral'))
chart.mark_circle(opacity=0.5, size=50).encode(color_encoding, 
                                               x=alt.X('x', axis=None), 
                                               y=alt.Y('y', axis=None))
```
